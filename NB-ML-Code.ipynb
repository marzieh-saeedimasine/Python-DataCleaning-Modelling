```python
## iris dataset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, make_blobs, make_moons, make_circles
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from mlxtend.plotting import plot_decision_regions

##iris
X=load_iris().data[:,[0,2]]
Y=load_iris().target
Xtrain, Xtest, Ytrain, Ytest=train_test_split(X,Y, test_size=0.3, random_state=1)

##train on X
model=GaussianNB()
model.fit(X,Y)
ypred=model.predict(X)
print(confusion_matrix(Y,ypred), accuracy_score(Y,ypred))

error=(ypred !=Y).sum()
ac=(Y.shape-error)/Y.shape
print(ac)

##train on xtrain
model.fit(Xtrain,Ytrain)
ypred=model.predict(Xtest)
print(confusion_matrix(Ytest,ypred), accuracy_score(Ytest,ypred))

##plot
plot_decision_regions(X,Y,clf=model)
```

```python
##make blobs

X,Y=make_blobs(n_samples=10,n_features=2, centers=2,random_state=10)
model=GaussianNB()
model.fit(X,Y)

## test dataset
print(X[:,0].min()-1, X[:,0].max()+1)
print(X[:,1].min()-1, X[:,1].max()+1)
rng=np.random.RandomState(0)
Xtest=[8,20]*rng.rand(2000,2)+[0,-12]
ypred=model.predict(Xtest)

## plot
plt.scatter(Xtest[:,0],Xtest[:,1], c=ypred, cmap='RdBu',s=20,alpha=0.1)
plt.scatter(X[:,0],X[:,1], c=Y, cmap='RdBu',s=50)
```
